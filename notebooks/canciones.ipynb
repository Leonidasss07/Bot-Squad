```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/canciones_populares.csv")

df["reproducciones"] = pd.to_numeric(df["reproducciones"], errors="coerce")
df = df.dropna(subset=["reproducciones"])

top10 = df.nlargest(10, "reproducciones")
top10["reproducciones_millones"] = top10["reproducciones"] / 1_000_000

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top10["nombre"], top10["reproducciones_millones"])

ax.set_title("Top canciones populares")
ax.set_xlabel("Reproducciones (millones)")
ax.set_ylabel("Canción")

ax.invert_yaxis()
plt.tight_layout()
plt.show()
```
