```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/generos_canciones.csv")

print(df.columns)
print(len(df))

conteo_generos = (
    df["generos"]
    .dropna()
    .astype(str)
    .str.split(r"\s*,\s*|\s*\|\s*")
    .explode()
    .str.strip()
)

conteo_generos = conteo_generos[conteo_generos != ""]

conteo_generos = conteo_generos.value_counts()

plt.figure(figsize=(10, 8))
conteo_generos.plot(kind="bar")
plt.title("Géneros musicales en las canciones")
plt.xlabel("Género")
plt.ylabel("Número de canciones")
plt.xticks(rotation=60, ha="right")
plt.tight_layout()
plt.show()
```
