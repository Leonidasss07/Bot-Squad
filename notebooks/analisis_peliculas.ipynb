```python
import pandas as pd

#with open(...)

df_series = pd.read_csv("../data/clean/generos_canciones.csv")

df_series.head(3)
```

```python
def separar_genero(genero):
    genero = genero.lower().strip()
    if 'rock' in genero:
        return 'rock'
    if 'pop' in genero:
        return 'pop'
    if 'hip' in genero or 'rap' in genero:
        return 'hip-hop'
    if 'electronic' in genero or 'house' in genero or 'techno' in genero or 'edm' in genero:
        return 'electronic'
    if 'metal' in genero:
        return 'metal'
    if 'indie' in genero:
        return 'indie'
    if 'alternative' in genero:
        return 'alternative'
    if 'r&b' in genero or 'rnb' in genero or 'soul' in genero:
        return 'r&b'
    return genero

generos_series = (df_series["generos"]
    .str.split(r" \| ")
    .explode()
    .str.strip()
    .map(separar_genero)
    .value_counts())

plt.figure(figsize=(4, 2))
generos_series.plot(kind="bar")
plt.title("Géneros - Canciones")
plt.show()
```
